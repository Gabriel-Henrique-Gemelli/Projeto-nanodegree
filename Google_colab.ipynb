```python
treino_path = "/content/dados_projeto_evasao_treino.csv"
teste_path = "/content/dados_projeto_evasao_teste.csv"

# Leitura dos arquivos CSV
df_treino = pd.read_csv(treino_path)
df_teste = pd.read_csv(teste_path)

# Verificar as primeiras linhas de cada dataset
df_treino.head(), df_teste.head()
```

```python
import numpy as np

# Copiar o dataframe para não alterar o original ainda
df_treino_com_target = df_treino.copy()

# Criar variável alvo 'evadiu' com base em ausência de acesso recente
# Critério: se o número de dias desde o último acesso for maior que 30 dias, considera-se evasão
df_treino_com_target['evadiu'] = np.where(
    df_treino_com_target['nr_dias_desde_ultimo_acesso'] > 30, 1, 0
)

# Verificar a distribuição da nova variável alvo
df_treino_com_target['evadiu'].value_counts()

# esse e o primeiro dia trabalhando no projeto 👍👍👍😒😒😍😍👌👌😘😘
```
